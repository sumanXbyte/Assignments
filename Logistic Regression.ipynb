1: What is Simple Linear Regression (SLR)? Explain its purpose.
-> Simple Linear Regression (SLR) is a supervised machine learning algorithm used to model the relationship between one independent variable (X) and one dependent variable (Y) using a straight line.

Its purpose is to:

1.Predict the value of the dependent variable
2.Understand the relationship between variables
3.Analyze how changes in X affect Y

2: What are the key assumptions of Simple Linear Regression?
-> The key assumptions of Simple Linear Regression are:

Linearity, Independence, Homoscedasticity, Normality, No multicollinearity

->  Y=β0​+β1​X+ε

4: Provide a real-world example where simple linear regression can be applied.
-> A real-world example is predicting house price based on house area.

5: What is the method of least squares in linear regression?
-> The method of least squares is a technique used to find the best-fitting line by minimizing the sum of squared errors.

6: What is Logistic Regression? How does it differ from Linear Regression?
-> Logistic Regression is used for classification problems, while Linear Regression is used for prediction of continuous values.

7: Name and briefly describe three common evaluation metrics for regression models.
-> Mean Absolute Error (MAE), Mean Squared Error (MSE), Root Mean Squared Error (RMSE)

8: What is the purpose of the R-squared metric in regression analysis?
-> R² measures how well the model explains the variability of the dependent variable.


```python
# 9: Write Python code to fit a simple linear regression model using scikit-learn.

from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
X = np.array([[1], [2], [3], [4]])
y = np.array([2, 4, 6, 8])

# Model
model = LinearRegression()
model.fit(X, y)

# Output
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```

10: How do you interpret the coefficients in a simple linear regression model?
-> The coefficient represents the change in output for one unit change in input.
